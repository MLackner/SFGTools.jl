```julia
get_attribute(scan[1], "comment")
```

```julia
dlpos = get_attribute(scan, "micos_position_x")
dlpos = (dlpos - 5000) * -1 *1e-6 / 3e8 * 1e12
```

```julia
z = zeros(size(scan[1].s, 1), length(scan))
for i = 1:length(scan)
    z[:,i] = squeeze(scan[i].s, 2)
end
contourf(dlpos, wavelength, z, cmap=ColorMap("magma"))
xlabel("Time Delay Upconversion [ps]")
ylabel("IR Wavelength")
```

```julia
using Loess
zz = similar(z)
for i = 1:size(z, 2)
    model = loess(collect(1.0:size(z, 1)), z[:,i], span=0.08, normalize=true)
    zz[:,i] = Loess.predict(model, collect(1.0:size(z, 1)))
end
```

```julia
figure()
contourf(dlpos, wavelength, zz, cmap=ColorMap("magma"), 
        levels=linspace(minimum(zz), maximum(zz), 20),
        linewidths=0.8)
colorbar(label="Counts per Second")
xlabel("Time Delay Upconversion [ps]")
ylabel("IR Wavelength [nm]")
```

```julia
pidx = [3, 23, 4, 21, 5, 16, 6, 9, 7]
c = ["tab:blue", "tab:orange", 
    "tab:blue", "tab:orange",
    "tab:blue", "tab:orange",
    "tab:blue", "tab:orange", "k"]
ax = figure()
for (i, pi) in enumerate(pidx)
    p1 = plot(wavelength, zz[:,pi], color=c[i], alpha=0.8, linestyle="-")
    p2 = plot(wavelength, z[:,pi], color=c[i], ".", alpha=0.2)
end
xlabel("IR Wavelength [nm]")
ylabel("Counts per Second")
legend(["Positive Delay", "Data", "Negative Delay", "Data"])
```
